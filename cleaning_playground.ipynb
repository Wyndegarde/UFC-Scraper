```python
percent_stats = [
            column.replace("red_", "").replace("blue_", "")
            for column in clean_ufc_df.columns
            if "percent" in column
        ]
percent_stats = list(set(percent_stats))
percent_stats
```

```python
ordered_stats = {
    "sig_str_percent": [],
    "total_str_percent": [],
    "td_percent": [],
    "sig_strike_defence_percent": [],
    "td_defence_percent": [],
}
```

```python
'''
- for each fighter, create the df, then iterate through the list of tuples.

'''
```

```python
single_fighter_df = clean_ufc_df[(clean_ufc_df["red_fighter"] == "Conor McGregor") | (clean_ufc_df["blue_fighter"] == "Conor McGregor")]
```

```python
ordered_stats = {
    "sig_str_percent": [],
    "total_str_percent": [],
    "td_percent": [],
    "sig_strike_defence_percent": [],
    "td_defence_percent": [],
}

for index, row in single_fighter_df.iterrows():
    if row["red_fighter"] == "Conor McGregor":
        for stat in percent_stats:
            ordered_stats[stat].append(row["red_" + stat])
    elif row["blue_fighter"] == "Conor McGregor":
        for stat in percent_stats:
            ordered_stats[stat].append(row["blue_" + stat])
```

```python
ordered_stats = defaultdict(list)

single_fighter_df = clean_ufc_df[(clean_ufc_df["red_fighter"] == "Conor McGregor") | (clean_ufc_df["blue_fighter"] == "Conor McGregor")]
for index, row in single_fighter_df.iterrows():
    if row["red_fighter"] == "Conor McGregor":
        for stat in percent_stats:
            ordered_stats[stat].append(row["red_" + stat])
    elif row["blue_fighter"] == "Conor McGregor":
        for stat in percent_stats:
            ordered_stats[stat].append(row["blue_" + stat])
```
